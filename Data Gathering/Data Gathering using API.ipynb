```python
# Task: Multiple pages/url's API data fetching
df = pd.DataFrame()
city = ['Goa', 'Mumbai', 'Chennai', 'Delhi', 'Kolkata', 'Bengaluru', 'Pune', 'Nashik', 'Amritsar', 'Ranchi', 'Patna', 'Ahmadabad ', 'Jaipur']
for i in city:
    url = 'https://api.openweathermap.org/data/2.5/weather?q={}&appid=dafd34e94fe728ccacca07d92ec25028&units=metric'.format(i)
    response1= requests.get(url)
    temp_df = pd.DataFrame(response1.json()['weather'])
    df = df.append(temp_df, ignore_index = True)
df.to_csv('cities.csv')
df
    
```
