```python
# sep & names parameter
# If column names are not given, use names parameter to give names
pd.read_csv('cities10.tsv', sep='\t', names=['ac','city','amount','dollar'])
```

```python
# Index col parameter: Use any column as index column
pd.read_csv('C:/Users/dell/Documents/desktop folders/Algo Practice/height-weight.csv', index_col='Weight')
```

```python
# Header parameter: If column names act as 1st row, use it to make them headers
pd.read_csv('cities10.tsv', sep='\t', header=1)
```

```python
# usecols parameter

pd.read_csv('cities10.tsv', sep='\t', names=['ac','city','amount','dollar'], usecols=['ac','city'])
```

```python
# squeeze parameter: If we want single column, convert it into pandas series instead of df
pd.read_csv('cities10.tsv', sep='\t', names=['ac','city','amount','dollar'], usecols=['ac'], squeeze=True)
```

```python
# skiprows/nrows parameter: skiprows skips unwanted index rows , nrows gets only specified no. of rows 
pd.read_csv('C:/Users/dell/Documents/desktop folders/Algo Practice/height-weight.csv', skiprows=[0,1], nrows=7)
```

```python
# Converters 
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
rename("Royal Challengers Bangalore")
```

```python
pd.read_csv('matches.csv', converters = {'team1':rename})
```
