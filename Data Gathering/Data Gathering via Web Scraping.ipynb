```python
# Code for whole dataset: 
final = pd.DataFrame()
for b in range(1,51):
    url = "https://books.toscrape.com/catalogue/page-{}.html".format(b)
    webpage = requests.get(url).text
    soup = BeautifulSoup(webpage,'lxml')
    books = soup.find_all('li', class_="col-xs-6 col-sm-4 col-md-3 col-lg-3")
    name = []
    price = []
    stock = []
    for book in books:
        name.append(book.find('h3').text.strip())
        str = book.find('p', class_="price_color").text.strip()
        s = slice(2,7)
        price.append(str[s])
        stock.append(book.find('p', class_="instock availability").text.strip())
    d = {'Name':name, 'Price':price, 'Stock':stock}
    df = pd.DataFrame(d)
    final = final.append(df, ignore_index=True)
print(final)
final.to_excel("Books.xlsx")
```
